```python
#create bins for decades
crash_bins = [1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]

# Create the names for the decade bins
crashes_by_decade = ["1900-1909", "1910-1919", "1920-1929", "1930-1939", "1940-1949", "1950-1959", "1960-1969", "1970-1979", "1980-1989", "1990-1999", "2000-2009", "2010-2020"]
                
# sort crashes into bins 
flight_df["Crashes by Decade"] = pd.cut(flight_df["Year"],crash_bins, labels=crashes_by_decade)
flight_df
```
